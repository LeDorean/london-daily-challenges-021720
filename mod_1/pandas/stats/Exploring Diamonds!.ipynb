**BONUS 1: Use the seaborn library to make a scatter plot between carat and depth, carat and table, carat and price. Use the hue argument to group by either color or cut**

```python
# Your code here! 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.6)
k_depth = sns.scatterplot(x="carat", y="depth",

                     hue="cut", data=df)
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.6)
k_table = sns.scatterplot(x="carat", y="table",

                     hue="cut", data=df)
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.6)
k_price = sns.scatterplot(x="price", y="carat",

                     hue="cut", data=df)
```
