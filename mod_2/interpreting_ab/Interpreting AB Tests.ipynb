```python
import numpy as np
import pandas as pd
import statsmodels as sm
from statsmodels.stats.power import TTestIndPower
from statsmodels.stats.weightstats import ttest_ind
import math

df = pd.read_csv("experiment_data_small.csv")
df
```

```python
mu_1 = df['baseline_small'].mean()
mu_2 = df['experiment_small'].mean()

sigma_1 = df['baseline_small'].std()
sigma_2 = df['experiment_small'].std()

pool_sigma = math.sqrt(((sigma_1**2 + sigma_2**2)/2))

pool_sigma
```

```python
def cohen_d(mean_1, mean_2, pooled_sd):
    diff = (mean_1 - mean_2)
    cohen_d = diff/pooled_sd
    return(cohen_d)

cohen_d(mu_1, mu_2, pool_sigma)
```

```python
# Set Parameters 


effect = .45
alpha = .05
power = .9

# Perform Power Analysis
analysis = TTestIndPower()
result = analysis.solve_power(effect, power=power, nobs1=None, ratio=1.0, alpha=alpha)
print('Sample Size: %.3f' % result)
```

```python
big_test = sm.stats.weightstats.ttest_ind(df['baseline_small'], 
                               df['experiment_small'], 
                               alternative='two-sided', 
                               usevar='pooled', value=0)
                                                          
                                                          
print("The critical value is:", big_test[0])
print("The p value value is:", big_test[1])                                                          
print("Degrees of Freedom:", big_test[2])                                                          
```
