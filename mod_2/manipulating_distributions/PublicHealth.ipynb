```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mu, sigma = 0, 1
n = 20000
bell_1 = np.random.normal(mu, sigma/3, n)
bell_2 = np.random.normal(mu+(sigma), sigma, n)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

sns.distplot(bell_1, hist=False, color='lightcoral', kde_kws={"shade": True})
sns.distplot(bell_2, hist=False, color='grey', kde_kws={"shade": True})

plt.xlabel("Time since first case")
plt.ylabel("daily number of cases", )
plt.title("Flattening The Curve", loc='left');
plt.xticks([])
plt.yticks([])

ax.axhline(y=0.4, ls='--', color='grey')
plt.annotate('Cases without protective measures',(-1,0.7))
plt.annotate('Cases with protective measures',(-.1,0.15))
plt.annotate('Health care system capacity',(1,0.43))
```
